```python
import pygal
```

```python
year_data = ['Easy-Task', 'Diff-Task']

garph = pygal.StackedBar()    # 创建图（叠加柱状图）

correct_data = [173, 131]
wrong_data = [2, 44]
# 添加数据
garph.add('Correct answer', correct_data)
garph.add('Wrong answer', wrong_data)

garph.x_labels = year_data            # 设置 X 轴刻度
garph.y_label_rotation = 45           # 设置Y轴的标签旋转多少度

garph.title = 'The ratio of correct answer versus wrong answer'  # 设置图标题
garph.x_title = 'Task-difficulty'                 # 设置 X 轴标题
garph.y_title = 'Count'            # 设置 Y 轴标题


garph.legend_at_bottom = True         # 设置图例位置（下面）

garph.margin = 35    # 设置页边距（margin_bottom、margin_top、margin_left、margin_right）

garph.show_x_guides = True    # 显示X轴的网格线
garph.show_y_guides = True    # 显示Y轴的网格线
```

```python
garph
```

```python
line_chart = pygal.Line()
line_chart.title = 'Ratio of correct answer (in %)'
line_chart.x_labels = ['session1', 'session2', 'session3', 'session4']
line_chart.add('Easy-Task', [100, 97.8, 97.6, 100])
line_chart.add('Diff-Task',  [62.8, 81.8, 68.18, 86.4])

```
