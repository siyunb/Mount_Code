```python
from nipype import Node, Workflow, Merge, IdentityInterface, SelectFiles, MapNode
from nipype.interfaces.io import DataSink
from nipype.interfaces.fsl import ExtractROI, MCFLIRT, FLIRT, Threshold, ApplyMask, Merge
```

```python
subject_list = [['%02d'%i] for i in range(12,23)]
```

```python
for subject in subject_list:
    # String template with {}-based strings
    templates = {'func': 'sub-{subject_id}/wsub-{subject_id}_ses-test_session*_bold.nii'}
    # Create SelectFiles node
    selectfiles = Node(SelectFiles(templates,
                            base_directory='/home/neuro/Result/Visual_search_task/1st_analysis/analysis_result/1st_normalized/datasink_spm/norm_spm',
                            sort_filelist=True), name='selectfiles')
    selectfiles.iterables = [('subject_id', subject)]

    merger = Node(Merge(dimension='t', output_type='NIFTI_GZ', tr=2), name='merger')

    output_folder = 'datasink_spm'
    datasink = Node(DataSink(base_directory='/home/neuro/Result/Visual_search_task/1st_analysis/analysis_result/1st_normalized/',
                            container=output_folder), name="datasink")
    substitutions = [('_subject_id_', 'sub-')]
    datasink.inputs.substitutions = substitutions

    mergeflow = Workflow(base_dir='/home/neuro/Result/Visual_search_task/preproc_working_dir', name='mergeflow')
    mergeflow.connect([(selectfiles, merger,[('func', 'in_files')]),
                    (merger, datasink,[('merged_file','norm_spm.@merge')])])

    mergeflow.run()
```
